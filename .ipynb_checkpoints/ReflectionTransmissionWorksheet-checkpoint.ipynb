```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def R_p (n1,n2,theta):
    if n1*np.sin(theta)/n2 >= 1:
        return 1
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = n2*np.cos(theta) - n1*np.cos(theta_t)
    B = n2*np.cos (theta) + n1*np.cos(theta_t)
    return (A/B)
```

```python
angle = np.linspace(0,np.pi/2,90)
```

```python
def T_p (n1,n2,theta):
    if n1*np.sin(theta)/n2 >= 1:
        return 1
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = 2*n1*np.cos(theta)
    B = n2*np.cos (theta) + n1*np.cos(theta_t)
    return (A/B)
```

```python
r_p = [R_p(1,1.5,x) for x in angle]
t_p = [T_p(1,1.5,x) for x in angle]
```

```python
plt.plot(angle,r_p)
plt.plot(angle,t_p)
```

```python
def R_s (n1,n2,theta):
    if n1*np.sin(theta)/n2 >= 1:
        return 1
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = n1*np.cos(theta) - n2*np.cos(theta_t)
    B = n1*np.cos (theta) + n2*np.cos(theta_t)
    return (A/B)
```

```python
def T_s (n1,n2,theta):
    if n1*np.sin(theta)/n2 >= 1:
        return 1
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = 2*n1*np.cos(theta)
    B = n1*np.cos (theta) + n2*np.cos(theta_t)
    return (A/B)
```

```python
r_s = [R_s(1,1.5,x) for x in angle]
t_s = [T_s(1,1.5,x) for x in angle]
```

```python
plt.plot(angle,r_s)
plt.plot(angle,t_s)
```

```python
#Same function but with light passing from air to glass (n1=1.5 and n2=1)
r_p2 = [R_p(1.5,1,x) for x in angle]
t_p2 = [T_p(1.5,1,x) for x in angle]

r_s2 = [R_s(1.5,1,x) for x in angle]
t_s2 = [T_s(1.5,1,x) for x in angle]
```

```python
#Reflectivity

def reflectivity (r_p):
    R = [r**2 for r in r_p]
    return R
```

```python
#transmittivity functions

#Transmittivity of parallel rays
def transmittivity_p (n1,n2,theta):
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = 2*n1*np.cos(theta)
    B = n2*np.cos (theta) + n1*np.cos(theta_t)
    t_p = A/B
    T = n2*(np.cos(theta_t)/np.cos(theta))*t_p**2
    return T

#Transmittivity of perpendicular rays
def transmittivity_s (n1,n2,theta):
    theta_t = np.arcsin(n1*np.sin(theta)/n2)
    A = 2*n1*np.cos(theta)
    B = n1*np.cos (theta) + n2*np.cos(theta_t)
    t_p = A/B
    T = n2*(np.cos(theta_t)/np.cos(theta))*t_p**2
    return T
```

```python
R_p = reflectivity(r_p)
T_p = [transmittivity_p(1,1.5,x) for x in angle]

plt.plot(angle,R_p)
plt.plot(angle,T_p)
```

```python
R_s = reflectivity(r_s)
T_s = [transmittivity_s(1,1.5,x) for x in angle]


plt.plot(angle,R_s)
plt.plot(angle,T_s)
```
