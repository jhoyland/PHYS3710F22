```python
import cmath
import numpy as np
from matplotlib import pyplot as plt
import
```

```python
N = 1e28
omega0 = 1e16
damp = 1e15
qe = 1.672e-19
epsil0 = 8.85e-12
me = 9.109e-31
```

```python
w1 = np.linspace(1e15,0.2e17,1000)
```

```python
def index(w):
    n = cmath.sqrt(1+(N*qe**2/(me*epsil0))/(omega0**2 - w**2 - 1j*damp*w))
    return (n)
```

```python
nr = [index(x).real for x in w1]
ni = [index(x).imag for x in w1]
```

```python
plt.plot(w1, nr)
plt.plot(w1, ni)
```
