```python
import numpy as np
import cmath as cm
import scipy.constants as sc
```

```python
def index(w):
    A = (1e28)*sc.e**2/(sc.electron_mass*sc.epsilon_0)
    B = 1.0/(1e16**2 - w**2 - 1j*1e15*w)
    n = cm.sqrt(1+A*B)
    return(n)
```

```python
w = np.linspace(1e15,0.2e17,1000)
n = [index(x).real for x in w]
```

```python
ni = [index(x).imag for x in w]
```

```python
from matplotlib import pyplot as plt
```

```python
index(1e16)
```

```python
plt.plot(w,n)
plt.plot(w,ni)
plt.show
```
