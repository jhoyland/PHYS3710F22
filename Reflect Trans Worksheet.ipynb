```python
import math as m
from matplotlib import pyplot as plt
import numpy as np
```

```python
def theta_t(angle,n1,n2):
    theta = m.asin(n1/n2*m.sin(angle))
    return theta

def r_p(angle,n1,n2):
    reflect = (n2*m.cos(angle))-n1*m.cos(theta_t(angle,n1,n2))/(n1*m.cos(theta_t(angle,n1,n2))+n2*m.cos(angle))
    return reflect

def t_p(angle,n1,n2):
    trans = (2*n1*m.cos(angle))/(n1*m.cos(theta_t(angle,n1,n2))+n2*m.cos(angle))
    return trans

def r_s(angle,n1,n2):
    reflect = (n1*m.cos(angle)-n2*m.cos(theta_t(angle,n1,n2)))/(n1*m.cos(angle)+n2*m.cos(theta_t(angle,n1,n2)))
    return reflect

def t_s(angle,n1,n2):
    trans = (2*n1*m.cos(angle))/(n2*m.cos(theta_t(angle,n1,n2)+n1*m.cos(angle)))
    return trans
```

Graphs of Reflection and Tranmission Coefficients - Low to High Refractive Indices

Reflection Coefficient - P-Polarized Light vs S-Polarized light


```python
n1_1 = 1
n2_1 = 1.5
x1 = np.linspace(0, m.pi/2, 100)
```

```python
y_rp1 = [r_p(xn,n1_1,n2_1) for xn in x1]
y_rs1 = [r_s(xn,n1_1,n2_1) for xn in x1]
plt.plot(x1,y_rp1)
plt.plot(x1,y_rs1)
```

Reflected Intensity

```python
y_Irp1 = [(r_p(xn,n1_1,n2_1))**2 for xn in x1]
y_Irs1 = [(r_s(xn,n1_1,n2_1))**2 for xn in x1]
plt.plot(x1,y_Irp1)
plt.plot(x1,y_Irs1)
```

Transmission Coefficient - P-Polarized Light vs S-Polarized light

```python
y_tp1 = [t_p(xn,n1_1,n2_1) for xn in x1]
y_ts1 = [t_s(xn,n1_1,n2_1) for xn in x1]
plt.plot(x1,y_tp1)
plt.plot(x1,y_ts1)
```

Transmission Intensity

```python
y_Itp1 = [(t_p(xn,n1_1,n2_1))**2 for xn in x1]
y_Its1 = [(t_s(xn,n1_1,n2_1))**2 for xn in x1]
plt.plot(x1,y_Itp1)
plt.plot(x1,y_Its1)
```
