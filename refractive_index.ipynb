```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

N = 1E28
w_o = 1E16
q_e = 1.602E-19
gamma = 1E15
m = 9.109E-31
e_o = 8.854E-12

A = (N*q_e**2)/((m*e_o))
```

```python
def r_index(w):
    n = cmath.sqrt(1+A*(1/(w_o**2-w**2-1j*gamma*w)))
    return n
```

```python
w = np.linspace(.7E16, 1.6E16, 1000)
n_real = [r_index(x).real for x in w]
n_imag = [r_index(x).imag for x in w]
```

```python
plt.plot(w,n_real)
plt.plot(w,n_imag)
```
