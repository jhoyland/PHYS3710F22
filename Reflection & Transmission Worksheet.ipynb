```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
theta = np.linspace(0,0.5*math.pi,180)
```

```python
nAir = 1
nGlass = 1.5
```

## Plotting rp, tp, rs, ts | going from air to glass

```python
rp1 = (nGlass*np.cos(theta) - nAir*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))/(nGlass*np.cos(theta) + nAir*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))
```

```python
plt.plot(theta,rp1)
```

```python
tp1 = 2*nAir*np.cos(theta) / (nGlass*np.cos(theta)+nAir*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))
```

```python
plt.plot(theta,tp1)
```

```python
rs1 = (nAir*np.cos(theta) - nGlass*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))/(nAir*np.cos(theta) + nGlass*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))
```

```python
plt.plot(theta,rs1)
```

```python
ts1 = 2*nAir*np.cos(theta) / (nAir*np.cos(theta)+nGlass*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2))
```

```python
plt.plot(theta,ts1)
```

## Plotting rp, tp, rs, ts | going from glass to air

```python
thetacrit = np.linspace(0,np.arcsin(2/3),180)
```

```python
rp2 = (nAir*np.cos(thetacrit) - nGlass*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))/(nAir*np.cos(thetacrit) + nGlass*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))
```

```python
plt.plot(thetacrit,rp2)
```

```python
tp2 = 2*nGlass*np.cos(thetacrit) / (nAir*np.cos(thetacrit)+nGlass*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))
```

```python
plt.plot(thetacrit,tp2)
```

```python
rs2 = (nGlass*np.cos(thetacrit) - nAir*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))/(nGlass*np.cos(thetacrit) + nAir*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))
```

```python
plt.plot(thetacrit,rp2)
```

```python
ts2 = 2*nGlass*np.cos(thetacrit) / (nGlass*np.cos(thetacrit)+nAir*np.sqrt(1-(nGlass*np.sin(thetacrit)/nAir)**2))
```

```python
plt.plot(thetacrit,tp2)
```

## Plotting $R_{p}, R_{s}, T_{p}$ and $T_{s}$, from air to glass

```python
Rp1 = rp1**2
plt.plot(theta,Rp1)
```

```python
Rs1 = rs1**2
plt.plot(theta,Rs1)
```

```python
Tp1 = nGlass*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2)/(nAir*np.cos(theta))*tp1**2
plt.plot(theta,Tp1)
```

```python
Ts1 = nGlass*np.sqrt(1-(nAir*np.sin(theta)/nGlass)**2)/(nAir*np.cos(theta))*ts1**2
plt.plot(theta,Ts1)
```
